```
import pandas as pd
```

```
df = pd.read_excel('product-usage.xlsx')
df.head()
```

```
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
df['DayofWeek'] = df['TransactionDate'].apply(lambda time: time.weekday_name)
zf = df.groupby(['Name'])['DayofWeek'].agg(lambda x:x.value_counts().index[0]).reset_index()
zf
```

```
nf = df[['Name','ProductKey', 'Description1', 'QuantityOrdered']]
cf = nf.groupby(['Name', 'Description1'])['QuantityOrdered'].sum().reset_index()
bf = cf.loc[cf.groupby(['Name'])['QuantityOrdered'].idxmax()]     
yf = bf.rename(columns = {'Description1':'most ordered product', 'QuantityOrdered' : 'most ordered quantity'})
yf
```

```
jf = cf.drop(cf.groupby(['Name'])['QuantityOrdered'].idxmax())
sf = jf.loc[jf.groupby(['Name'])['QuantityOrdered'].idxmax()]     
xf = sf.rename(columns = {'Description1':'second most ordered product', 'QuantityOrdered' : 'second  ordered quantity'})
xf
```

```
k = pd.merge(yf, xf, on='Name')
k = pd.merge(k, zf, on='Name')
k
```

```
k.to_csv('output.csv', header=True)
from google.colab import files
files.download('output.csv')
```
